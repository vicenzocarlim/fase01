```python
# Importação de biblioteca necessária

import pandas as pd

# Importação da base de dados

df = pd.read_csv("https://raw.githubusercontent.com/vicenzocarlim/fase01/main/fase_01.csv")
```

```python
# Exibir as cinco primeiras linhas do DataFrame
df.head()
```


# 1. Tratamento dos dados

**QUESTÃO 01:** Trate os dados.

**1.** Tratar os valores NaN's

```python
# Conferir se existem valores NaN em cada coluna
df.isnull().sum()
```

Pode-se observar que todas as colunas que possuem valores NaN referem-se a métricas/números/quantidades. Portanto, substituiremos esses valores por 0. 

```python
# Trocar NaN's por 0
df.fillna(0, inplace = True)
```

**2.** Tratar a coluna 2, que corresponde à variável Data

```python
# Verificar se há alguma data destoante das demais com base no ano
# Para tal, colocar as datas no tipo 'datetime'

import datetime

df["Data"] = pd.to_datetime(df.Data)
# Exibir as 5 primeiras linhas para verificar a troca do formato da data
df.head() 
```

```python
# Exibir todos os anos que as postagens contemplam

df_serie = pd.Series(df["Data"])
df_serie_rem = df_serie.dt.year.sort_values().drop_duplicates()
df_serie_rem
```

```python
# Exibir as datas cujos anos não correspondem à realidade

df_2017 = df.loc[df["Data"].dt.year == 2017]
print(df_2017["Nome do Post"])
print(df_2017["Data"])
df_2023 = df.loc[df["Data"].dt.year == 2023]
print(df_2023["Nome do Post"])
print(df_2023["Data"])
df_2024 = df.loc[df["Data"].dt.year == 2024]
print(df_2024["Nome do Post"])
print(df_2024["Data"])
df_2025 = df.loc[df["Data"].dt.year == 2025]
print(df_2025["Nome do Post"])
print(df_2025["Data"])
```

```python
# Trocar os anos incorretos/destoantes pela data correta
# de acordo com o nome do post

df = df.replace(df["Data"][140], "2021-08-28")
df = df.replace(df["Data"][181], "2021-09-13")
df = df.replace(df["Data"][165], "2021-10-06")
df = df.replace(df["Data"][19], "2021-07-17")
df = df.replace(df["Data"][69], "2021-07-05")
df = df.replace(df["Data"][199], "2021-11-21")

# Exibir todos os anos que as postagens contemplam, a partir de agora

df_serie_2 = pd.Series(df["Data"])
df_serie_rem_2 = df_serie_2.dt.year.sort_values().drop_duplicates()
df_serie_rem_2
```

```python
# Voltar a data para o formato DD/MM/AAAA
df["Data"] = pd.to_datetime(df.Data)
df["Data"] = df["Data"].dt.strftime("%d/%m/%Y")
# Exibir as 5 primeiras linhas para verificar a troca do formato da data
df.head()
```

**3.** Tratar a coluna 4, que corresponde à variável Posicionamento

```python
# Descobrir quais são os tipos de posicionamento possíveis
# para, posteriormente, normalizá-los

df_serie = pd.Series(df["Posicionamento"])
df_serie_rem = df_serie.sort_values().drop_duplicates()
df_serie_rem
```

```python
# Normalizar os termos que correspondem aos tipos de posicionamentos

df = df.replace("Fed", "Feed")
df = df.replace("feed", "Feed")
df = df.replace("Storie", "Stories")
df = df.replace("Story", "Stories")
df = df.replace("stories", "Stories")

# Conferir se, de fato, houve a normalização dos termos

df.groupby("Posicionamento").sum()
```

**4.** Tratar a coluna 5, que corresponde à variável plataforma

```python
# Descobrir quais são os tipos de plataforma possíveis
# para, posteriormente, normalizá-los

df_serie = pd.Series(df["Plataforma"])
df_serie_rem = df_serie.sort_values().drop_duplicates()
df_serie_rem
```

```python
# Normalizar os termos que correspondem aos tipos de plataformas

df = df.replace("Linkedin", "LinkedIn")
df = df.replace("ig", "Instagram")

# Conferir se, de fato, houve a normalização dos termos

df.groupby("Plataforma").sum()
```

**5.** Tratar a coluna 6, que corresponde à variável Formato

```python
# Descobrir quais são os tipos de formato possíveis
# para, posteriormente, normalizá-los

df_serie = pd.Series(df["Formato"])
df_serie_rem = df_serie.sort_values().drop_duplicates()
df_serie_rem
```

```python
# Normalizar os termos que correspondem aos tipos de formato

df = df.replace("Photo", "Imagem")
df = df.replace("Video", "Vídeo")
df = df.replace("imagem", "Imagem")
df = df.replace("vídeo", "Vídeo")

# Conferir se, de fato, houve a normalização dos termos

df.groupby("Formato").sum()
```

**QUESTÃO 13:** Qual a média de curtidas no *Instagram*? E no *LinkedIn*?
> **Resposta:** A média de curtidas no *Instagram* é de 496,42 curtidas e, no *LinkedIn*, é de 94,913 curtidas, aproximadamente.
